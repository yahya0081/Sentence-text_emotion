```python
import pandas as pd        # For loading and manipulating data
import numpy as np         # For mathematical operations
import matplotlib.pyplot as plt  # For plotting graphs
import seaborn as sns      # For advanced plotting

from sklearn.model_selection import train_test_split   # For splitting dataset
from sklearn.feature_extraction.text import TfidfVectorizer  # For converting text to numbers
from sklearn.naive_bayes import MultinomialNB   # The model weâ€™ll use
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score  # To measure how good our model is
```

```python
df = pd.read_csv("train.txt", sep=";", names=["text", "emotion"])
print(df.head())
```

```python
import re
import string

def clean_text(text):
    text = text.lower()  # convert to lowercase
    text = re.sub(r'\[.*?\]', '', text)  # remove things like [text]
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # remove links
    text = re.sub(r'<.*?>+', '', text)  # remove HTML tags
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # remove punctuation
    text = re.sub(r'\n', '', text)  # remove newline characters
    text = re.sub(r'\w*\d\w*', '', text)  # remove words with numbers (e.g. h4ppy)
    return text

df['clean_text'] = df['text'].apply(clean_text)
```

```python
X = df['clean_text']       # Input: cleaned text
y = df['emotion']          # Output: emotion label (e.g., joy, anger)

tfidf = TfidfVectorizer(max_features=5000)  # Convert text into numerical values (5000 most useful words)
X_vec = tfidf.fit_transform(X).toarray()    # Transform the input text into a matrix of numbers
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

```

```python
model = MultinomialNB()  # Naive Bayes is fast and works well with text
model.fit(X_train, y_train)  # Train the model with our text and labels
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues",
            xticklabels=model.classes_, yticklabels=model.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
def predict_mood(text):
    cleaned = clean_text(text)
    vec = tfidf.transform([cleaned]).toarray()
    pred = model.predict(vec)[0]
    return pred

# Try it
user_input = input("How are you feeling today? ")
print("Detected Mood:", predict_mood(user_input))

```
